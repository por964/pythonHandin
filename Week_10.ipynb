```python
from modules.text_comparer import TextComparer
import matplotlib.pyplot as plt

link_two = ['https://www.gutenberg.org/files/244/244-0.txt',
            'https://www.gutenberg.org/files/902/902-0.txt',
            'https://www.gutenberg.org/files/829/829-0.txt',
            'https://www.gutenberg.org/files/64773/64773-0.txt',
            'https://www.gutenberg.org/files/64777/64777-0.txt'
            ]

tc2 = TextComparer(link_two)

tc2.multi_download()

to_plot = {}

help_list = []

for f in tc2.filenames:
    help_list.append(tc.avg_vowels(f))

to_plot.update(help_list)

sorted_tuples = sorted(to_plot.items(), key=lambda item: item[1], reverse=True)
print(sorted_tuples)
sorted_dict = {k: v for k, v in sorted_tuples}

print(sorted_dict)

keys = sorted_dict.keys()
values = sorted_dict.values()

plt.bar(keys, values)
plt.show()
```
